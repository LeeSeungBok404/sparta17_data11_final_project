이전에 만들었던 유튜브_유입지수와 팬덤지수
- 분모 : 팔로워 
- 분자 : 외부 지표 (유튜브_구독자, 팬카페_가입자, X_팔로워)

-> 현재 플랫폼 체급이 작을수록 지수값이 커짐

하위 50% 가 모두 0이라서 상위 10%가 낮게 나옴

팔로워가 0인 사람들을 제외하고 거기서 보자 

```python
import pandas as pd
import numpy as np

# 1. 데이터 로드
df = pd.read_csv('final_softcone.csv')

def get_active_thresholds(df, column, percentiles):
    """
    특정 컬럼에서 0을 제외한 데이터만을 대상으로 상위 N% 임계값을 계산 
    """
    active_data = df[df[column] > 0][column]
    
    if active_data.empty:
        return [0] * len(percentiles)
    
    # 상위 N% 계산 후 반올림하여 정수로 변환
    results = active_data.quantile([1 - (p/100) for p in percentiles])
    return results.round().astype(int).values

# 2. 분석 설정
target_cols = ['유튜브_구독자', '팬카페_가입자', 'X_팔로워']
percents = [5, 10, 15]

# 3. 결과 계산 및 출력
print(f"{'지표':<15} | {'활동자 수':<8} | {'상위 5%':<10} | {'상위 10%':<10} | {'상위 15%':<10}")
print("-" * 70)

for col in target_cols:
    active_count = len(df[df[col] > 0])
    v5, v10, v15 = get_active_thresholds(df, col, percents)
    
    # :d 또는 :,d 를 사용하면 정수형으로 콤마까지 찍어서 출력 가능해
    print(f"{col:<15} | {active_count:<10,d} | {v5:<12,d} | {v10:<12,d} | {v15:<12,d}")

# 4. 변수 저장용 (나중에 필터링 조건으로 쓸 때)
yt_10 = int(df[df['유튜브_구독자'] > 0]['유튜브_구독자'].quantile(0.90))
cafe_10 = int(df[df['팬카페_가입자'] > 0]['팬카페_가입자'].quantile(0.90))
x_10 = int(df[df['X_팔로워'] > 0]['X_팔로워'].quantile(0.90))

print(f"\n상위 10%:")
print(f"유튜브: {yt_10:,}명 | 팬카페: {cafe_10:,}명 | X: {x_10:,}명")
```

```python
import pandas as pd

# 1. 데이터 불러오기 (파일명은 실제 파일명으로 수정해주세요)
df = pd.read_csv('final_softcone.csv')

# =====================================================================
# [조건 1] 외부 커뮤니티 체급 상위권 (절대 수치 TOP 100)
# =====================================================================
# 각 지표별 상위 100명의 인덱스 추출 (0인 데이터는 제외)
top100_x = df[df['X_팔로워'] > 0].nlargest(200, 'X_팔로워').index
top100_yt = df[df['유튜브_구독자'] > 0].nlargest(200, '유튜브_구독자').index
top100_cafe = df[df['팬카페_가입자'] > 0].nlargest(200, '팬카페_가입자').index

# 세 지표 중 하나라도 TOP 100에 속하는 인원 (합집합)
cond1_idx = top100_x.union(top100_yt).union(top100_cafe)
cond1_mask = df.index.isin(cond1_idx)

# =====================================================================
# [조건 2] 버츄얼 체급은 아직 신입일 것 (상한선 + 지수 TOP 100)
# =====================================================================
# 2-1. 상한선(Upper Limit) 설정: 플랫폼 팔로워가 전체 평균 이하일 것
#mean_followers = df['최고_팔로워'].mean()
#upper_bound_mask = df['최고_팔로워'] <= mean_followers

# 2-2. 지표 비율 기준 TOP 100 추출
top100_yt_index = df.nlargest(100, '유튜브_유입지수').index
top100_fandom_index = df.nlargest(100, '팬덤지수').index

# 두 지수 중 하나라도 TOP 100에 속하는 인원
cond2_idx = top100_yt_index.union(top100_fandom_index)
index_mask = df.index.isin(cond2_idx)

# 조건 2 최종 마스크 (평균 이하의 팔로워 AND 지수 TOP 100)
cond2_mask = index_mask

# =====================================================================
# [최종 추출] 세그먼트 5: 중고신입형
# =====================================================================
# 조건 1과 조건 2를 동시에 만족하는 교집합 데이터 추출
segment_5_df = df[cond1_mask & cond2_mask]

# =====================================================================
# 결과 확인 및 저장
# =====================================================================
print(f"[데이터 요약]")
print(f"1. 조건 1(외부 네임드) 통과자: {cond1_mask.sum()}명")
print(f"2. 조건 2(방송 신입 + 지수 상위) 통과자: {cond2_mask.sum()}명")
print("-" * 40)
print(f"최종 세그먼트 5 (중고신입형) 발굴 인원: {len(segment_5_df)}명")

# 추출된 데이터를 CSV 파일로 저장 (팀원 공유용, 한글 깨짐 방지 utf-8-sig)
segment_5_df.to_csv('segment5_used_newbie.csv', index=False, encoding='utf-8-sig')

# 결과 데이터프레임 미리보기 (상위 5명)
segment_5_df[['스트리머명', '최고_팔로워', 'X_팔로워', '유튜브_구독자', '팬카페_가입자', '유튜브_유입지수', '팬덤지수']].head(60)
```

조건2에서 팬심지수는 X팔로워랑 카페회원수를 같이봄


팬심지수(X&팬카페)에서 -> X 유입지수

$$ X\text{ 유입지수} = \frac{\log_{10}(X\_\text{팔로워} + 1)}{\log_{10}(\text{최고\_팔로워} + 1)} $$

```python
import pandas as pd
import numpy as np

# 1. 데이터 불러오기
df = pd.read_csv('final_softcone.csv')


# X_유입지수 생성 (분모 0 처리를 위해 log10 적용 시 +1)
df['X_유입지수'] = np.log10(df['X_팔로워'] + 1) / np.log10(df['최고_팔로워'] + 1)

# inf(무한대)를 0으로 치환 (최고_팔로워가 0인 경우 대비)
df['X_유입지수'] = df['X_유입지수'].replace([np.inf, -np.inf], np.nan).fillna(0)

# =====================================================================
# [조건 1] 외부 커뮤니티 체급 상위권 (절대 수치 TOP 100)
# =====================================================================
top100_x = df[df['X_팔로워'] > 0].nlargest(200, 'X_팔로워').index
top100_yt = df[df['유튜브_구독자'] > 0].nlargest(200, '유튜브_구독자').index

cond1_idx = top100_x.union(top100_yt)
cond1_mask = df.index.isin(cond1_idx)

# =====================================================================
# [조건 2] 버츄얼 체급은 아직 신입일 것 (상한선 + 지수 TOP 100)
# =====================================================================

#mean_followers = df[df['최고_팔로워'] > 0]['최고_팔로워'].mean()
#upper_bound_mask = df['최고_팔로워'] <= mean_followers

# 2-2. 지표 비율 기준 TOP 100 추출 (유튜브와 X의 유입지수 활용)
top100_yt_inflow = df.nlargest(200, '유튜브_유입지수').index
top100_x_inflow = df.nlargest(200, 'X_유입지수').index

cond2_idx = top100_yt_inflow.union(top100_x_inflow)
index_mask = df.index.isin(cond2_idx)

# 최종 조건 2: 지수 상위권이면서 + 동시에 방송 체급은 평균 이하일 것
cond2_mask = index_mask 


# [최종 추출] 세그먼트 5: 중고신입형

segment_5_df = df[cond1_mask & cond2_mask]


# 결과 확인 
print(f"[데이터 요약]")
print(f"1. 조건 1(외부 네임드) 통과자: {cond1_mask.sum()}명")
print(f"2. 조건 2(지수 상위 & 신입 체급) 통과자: {cond2_mask.sum()}명")
print("-" * 40)
print(f"최종 세그먼트 5 (중고신입형) 발굴 인원: {len(segment_5_df)}명")

# CSV 저장
#segment_5_df.to_csv('segment5_used_newbie_v2.csv', index=False, encoding='utf-8-sig')

display_cols = ['스트리머명', '최고_팔로워', 'X_팔로워', '유튜브_구독자', '유튜브_유입지수', 'X_유입지수']
(segment_5_df[display_cols].sort_values(by='X_유입지수', ascending=False).tail(50))
```
