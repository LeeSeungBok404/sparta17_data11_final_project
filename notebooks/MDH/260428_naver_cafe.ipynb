# 소프트콘 팔로워 수 보기

```python
import pandas as pd
import math

# 1. CSV 파일 읽기
df = pd.read_csv('final_softcone_streamers_V2.csv')

# 결측치가 있다면 0으로 채우기 (데이터 안전성을 위해 추가)
df['팔로워'] = df['팔로워'].fillna(0)

# 2. 팔로워 수 최댓값 확인
max_followers = df['팔로워'].max()
print(f"전체 스트리머 중 최대 팔로워 수: {max_followers:,.0f}명\n")

# 3. 500 단위로 구간(bins)과 라벨(labels) 만들기
# 최댓값을 500으로 나누어 올림 처리한 후 다시 500을 곱해 최대 구간의 끝값을 구합니다.
max_bin = math.ceil(max_followers / 500) * 500

# 구간: [0, 500, 1000, 1500, ...]
bins = [0] + list(range(500, max_bin + 1, 500))

# 라벨: ['0~500', '501~1000', '1001~1500', ...]
labels = []
for i in range(len(bins)-1):
    if i == 0:
        labels.append("0~500")
    else:
        labels.append(f"{bins[i]+1}~{bins[i+1]}")

# 4. 구간 나누기 (include_lowest=True를 설정해야 0명인 스트리머도 첫 구간에 포함됩니다)
df['팔로워_구간'] = pd.cut(df['팔로워'], bins=bins, labels=labels, include_lowest=True)

# 5. 구간별 스트리머 수 카운트 및 정렬
follower_counts = df['팔로워_구간'].value_counts().sort_index()

# 6. 결과 출력
print("[ 팔로워 수 500명 단위 분포 ]")
for section, count in follower_counts.items():
    # 해당 구간에 스트리머가 1명이라도 있는 경우만 출력 (0명인 구간도 보고 싶다면 if문 제거)
    if count > 0: 
        print(f"{section}명 : {count}명")
```

```python
import pandas as pd
import math

# 1. CSV 파일 읽기
df = pd.read_csv('final_softcone_streamers_V2.csv')

# 결측치가 있다면 0으로 채우기
df['팔로워'] = df['팔로워'].fillna(0)

# 2. 팔로워 수 최댓값 확인
max_followers = df['팔로워'].max()
print(f"전체 스트리머 중 최대 팔로워 수: {max_followers:,.0f}명\n")

# 3. 1000 단위로 구간(bins)과 라벨(labels) 만들기 (★ 500 -> 1000으로 수정됨)
max_bin = math.ceil(max_followers / 1000) * 1000

# 구간: [0, 1000, 2000, 3000, ...]
bins = [0] + list(range(1000, max_bin + 1, 1000))

# 라벨: ['0~1000', '1001~2000', '2001~3000', ...]
labels = []
for i in range(len(bins)-1):
    if i == 0:
        labels.append("0~1000")
    else:
        labels.append(f"{bins[i]+1}~{bins[i+1]}")

# 4. 구간 나누기
df['팔로워_구간'] = pd.cut(df['팔로워'], bins=bins, labels=labels, include_lowest=True)

# 5. 구간별 스트리머 수 카운트 및 정렬
follower_counts = df['팔로워_구간'].value_counts().sort_index()

# 6. 결과 출력
print("[ 팔로워 수 1000명 단위 분포 ]")
for section, count in follower_counts.items():
    if count > 0: 
        print(f"{section}명 : {count}명")

print(f"전체 스트리머 수: {len(df)}명")
```

```python
import pandas as pd
import time
from datetime import datetime
from pathlib import Path

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

INPUT_FILE = "input.csv"
OUTPUT_FILE = "fancafe_collected.csv"


def now_str():
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")


def is_missing(v):
    return pd.isna(v) or str(v).strip() == ""


def ensure_columns(df):
    cols = {
        "fancafe_url": "",
        "fancafe_member_count": "",
        "fancafe_post_count": "",
        "fancafe_status": "",
        "fancafe_collected_at": ""
    }
    for c in cols:
        if c not in df.columns:
            df[c] = ""
    return df


def needs_processing(row):
    if is_missing(row.get("fancafe_url")):
        return False
    if is_missing(row.get("fancafe_member_count")):
        return True
    return False


def crawl_cafe(driver, url):
    result = {
        "member": "",
        "post": "",
        "status": ""
    }

    try:
        driver.get(url)
        time.sleep(2)

        # 🔥 iframe 진입 (핵심)
        driver.switch_to.frame("cafe_main")
        time.sleep(1)

        # 회원 수
        try:
            member = driver.find_element(By.CSS_SELECTOR, ".cafe_info .num")
            result["member"] = member.text
        except:
            result["member"] = ""

        # 전체 글 수
        try:
            total = driver.find_element(By.XPATH, "//a[contains(text(),'전체보기')]")
            result["post"] = total.find_element(By.XPATH, "./following-sibling::span").text
        except:
            result["post"] = ""

        result["status"] = "OK"

    except Exception as e:
        result["status"] = f"ERROR: {e}"

    finally:
        driver.switch_to.default_content()

    return result


def main():
    input_path = Path(INPUT_FILE)
    output_path = Path(OUTPUT_FILE)

    if output_path.exists():
        print("👉 이어서 실행")
        df = pd.read_csv(output_path)
    else:
        print("👉 새로 시작")
        df = pd.read_csv(input_path)

    df = ensure_columns(df)

    targets = [idx for idx, row in df.iterrows() if needs_processing(row)]

    print("총 대상:", len(targets))

    options = Options()
    options.add_argument("start-maximized")
    driver = webdriver.Chrome(options=options)

    for i, idx in enumerate(targets, 1):
        row = df.loc[idx]
        url = row["fancafe_url"]

        print(f"\n[{i}/{len(targets)}] {url}")

        result = crawl_cafe(driver, url)

        df.at[idx, "fancafe_member_count"] = result["member"]
        df.at[idx, "fancafe_post_count"] = result["post"]
        df.at[idx, "fancafe_status"] = result["status"]
        df.at[idx, "fancafe_collected_at"] = now_str()

        print("회원:", result["member"])
        print("글:", result["post"])
        print("상태:", result["status"])

        # 🔥 중간 저장
        df.to_csv(output_path, index=False, encoding="utf-8-sig")

    driver.quit()

    print("\n완료!")


if __name__ == "__main__":
    main()
```
