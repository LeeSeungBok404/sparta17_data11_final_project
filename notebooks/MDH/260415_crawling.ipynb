## 1일 기준 스트리머 TOP 100 채널 아이디 가져오기

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

channel_info_list = []

try:
    # 페이지 접속
    driver.get("https://playboard.co/youtube-ranking/most-viewed-v-tuber-channels-in-south-korea-daily")
    time.sleep(5)
    
    for _ in range(10): 
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        
    # 1. 이름 요소 찾기 
    names_elements = driver.find_elements(By.XPATH, "//h3[@data-v-2ea9480b]")
    
    # 2. 채널 주소가 담긴 <a> 태그 찾기
    link_elements = driver.find_elements(By.XPATH, "//a[contains(@href, '/channel/UC')]")

    print(f"이름 {len(names_elements)}개, 링크 {len(link_elements)}개 발견!")

    # 3. 짝지어서 ID만 깔끔하게 잘라내기
    count = min(len(names_elements), len(link_elements))
    for i in range(count):
        name = names_elements[i].text
        
        # href 주소 가져오기 
        raw_url = link_elements[i].get_attribute("href")
        
        # 주소에서 'UC'로 시작하는 진짜 ID 부분만 뚝 떼어내기
        channel_id = raw_url.split("/channel/")[1] if "/channel/" in raw_url else "ID_오류"
        
        channel_info_list.append({
            "Name": name,
            "Channel ID": channel_id
        })
        print(f"{name} -> {channel_id}")

except Exception as e:
    print(f"에러 발생: {e}")

finally:
    driver.quit()
    
    # 4. '마스터 사전' 파일로 저장
    if channel_info_list:
        df_info = pd.DataFrame(channel_info_list)
        
        # 중복된 채널이 있다면 하나만 남기고 제거 (깔끔한 사전을 위해!)
        df_info = df_info.drop_duplicates(subset=['Name']) 
        
        df_info.to_csv("test2.csv", index=False, encoding="utf-8-sig")
        print("\n'channel_info.csv' 마스터 파일이 저장되었습니다!")
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import random

# 1. 크롬 옵션 설정 
chrome_options = Options()
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")
# 2. 드라이버 실행 
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

try:

    # 1단계: 3월 조회수 TOP 100 명단 수집
    period_code = "1774915200" 
    views_url = f"https://playboard.co/youtube-ranking/most-viewed-v-tuber-channels-in-south-korea-monthly?period={period_code}"

    print("\n3월 조회수 TOP 100 명단 수집 중")
    driver.get(views_url)
    time.sleep(30)

    for _ in range(4):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)

    names = driver.find_elements(By.XPATH, "//h3[@data-v-2ea9480b]")
    views = driver.find_elements(By.XPATH, "//span[@data-v-2ea9480b and contains(@class, 'fluc-label')]")
    links = driver.find_elements(By.XPATH, "//a[contains(@href, '/channel/UC')]")

    top_100_list = []
    count = min(100, len(names), len(views), len(links))

    for i in range(count):
        channel_id = links[i].get_attribute("href").split("/channel/")[1] if "/channel/" in links[i].get_attribute("href") else ""
        top_100_list.append({
            "순위": i + 1,
            "Channel ID": channel_id,
            "채널명": names[i].text,
            "조회수": views[i].text,
            "슈퍼챗": "0",
            "구독자": "0"
        })

    print(f"명단 수집 완료! (총 {len(top_100_list)}명)")

    # 2단계: 상세 페이지 방문
    print("\n버튜버 데이터 수집 시작...")
    time.sleep(30)


    # 테스트용 상위 3명 
    target_list = top_100_list[:2] 

    for idx, streamer in enumerate(target_list): 
        cid = streamer["Channel ID"]
        if not cid: continue

        print(f"[{idx+1}/{len(target_list)}] '{streamer['채널명']}' 수집 중...")

        # --- 슈퍼챗 수집 ---
        driver.get(f"https://playboard.co/channel/{cid}/superchat?period={period_code}")
        time.sleep(random.uniform(4, 6)) # 로봇 방지를 위해 랜덤하게 대기

        try:
            total_superchat = driver.find_element(By.XPATH, "//tr[td[contains(text(), '합계')]]/td[2]").text
            streamer["슈퍼챗"] = total_superchat
        except:
            streamer["슈퍼챗"] = "미집계" 

        # --- 구독자 수집 ---
        driver.get(f"https://playboard.co/channel/{cid}/subscribers?period={period_code}")
        time.sleep(random.uniform(4, 6))

        try:
            total_sub = driver.find_element(By.XPATH, "//tr[td[contains(text(), '합계')]]/td[2]").text
            streamer["구독자"] = total_sub
        except:
            streamer["구독자"] = "미집계"

    # 3. 데이터 저장
    df_final = pd.DataFrame(target_list)
    df_final.to_csv("homework.csv", index=False, encoding="utf-8-sig")
    print("\n모든 작업 완료")

except Exception as e:
    print(f"에러 발생: {e}")

finally:
    driver.quit()
```

# 2026년 3월 TOP 43 슈퍼챗 순위

### 순위, 채널명, 태그, 채널 ID, 유튜브 링크, 슈퍼챗 수입, 슈퍼챗 개수

```python
import pandas as pd
import re
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

# 1. 브라우저 설정
chrome_options = Options()
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

try:
    period_code = "1774915200"
    url = f"https://playboard.co/youtube-ranking/most-superchatted-v-tuber-channels-in-south-korea-monthly?period={period_code}"
    
    print(f"데이터를 수집합니다")
    driver.get(url)
    time.sleep(5)

    for i in range(5):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)

    # 행단위로 가져오기
    rows = driver.find_elements(By.XPATH, "//tr[descendant::h3]") 
    if not rows: 
        rows = driver.find_elements(By.XPATH, "//ul/li[descendant::h3]")

    result_list = []

    for i, row in enumerate(rows[:100]):
        try:
            # 1. 채널명
            name = row.find_element(By.XPATH, ".//h3").text
            
            # 2. 채널 ID 추출
            link_el = row.find_element(By.XPATH, ".//a[contains(@href, '/channel/UC')]")
            href = link_el.get_attribute("href")
            channel_id = href.split("/channel/")[1] if "/channel/" in href else "ID 찾음 실패"
            
            # 유튜브 바로가기 링크 생성
            youtube_link = f"https://youtube.com/channel/{channel_id}" if channel_id != "ID 찾음 실패" else ""

            # 해시태그(#) 수집
            tag_elements = row.find_elements(By.XPATH, ".//ul/li")
            tags = ", ".join([t.text for t in tag_elements if "#" in t.text])
            
            # 3. 슈퍼챗 수입 및 개수
            spans = row.find_elements(By.XPATH, ".//span[contains(@class, 'fluc-label')]")
            revenue = 0
            chat_count = 0
            
            if len(spans) >= 1:
                clean_rev = re.sub(r'[^0-9]', '', spans[0].text)
                revenue = int(clean_rev) if clean_rev else 0
            if len(spans) >= 2:
                clean_cnt = re.sub(r'[^0-9]', '', spans[1].text)
                chat_count = int(clean_cnt) if clean_cnt else 0

            
            result_list.append({
                "순위": i + 1,
                "채널명": name,
                "태그": tags,              
                "채널 ID": channel_id,
                "유튜브 링크": youtube_link, 
                "슈퍼챗 수입": revenue,
                "슈퍼챗 개수": chat_count
            })
            
        except Exception as e:
            print(f"{i+1}위 데이터 수집 중 누락 발생: {e}")
            continue

    # 데이터프레임 저장
    df = pd.DataFrame(result_list)
    filename = "v_tuber_superchat_march_2026_final.csv"
    df.to_csv(filename, index=False, encoding="utf-8-sig")
    
    print("-" * 30)
    print(f"수집 완료")
    print(f"파일명: {filename}")

except Exception as e:
    print(f"에러 발생: {e}")

finally:
    driver.quit()
```

```python
import pandas as pd
import re
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import datetime

# 1. 브라우저 설정
chrome_options = Options()
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

# 2. 수집할 월과 타임스탬프 생성
target_months = []
for year in [2025, 2026]:
    for month in range(1, 13):
        if year == 2026 and month > 3: # 2026년은 3월까지만
            break
        # UTC 기준 매월 1일의 타임스탬프 생성
        dt = datetime.datetime(year, month, 1, 0, 0, tzinfo=datetime.timezone.utc)
        target_months.append((f"{year}-{month:02d}", str(int(dt.timestamp()))))

result_list = []

try:
    # 3. 월별 데이터 수집 루프
    for date_str, period_code in target_months:
        url = f"https://playboard.co/youtube-ranking/most-superchatted-v-tuber-channels-in-south-korea-monthly?period={period_code}"
        
        print(f"[{date_str}] 데이터를 수집중... (URL: {url})")
        driver.get(url)
        time.sleep(15) # 페이지 로딩 대기

        # 스크롤 내리기 (숨겨진 100위까지 로딩)
        for _ in range(5):
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)

        # 행 단위로 가져오기
        rows = driver.find_elements(By.XPATH, "//tr[descendant::h3]") 
        if not rows: 
            rows = driver.find_elements(By.XPATH, "//ul/li[descendant::h3]")

        for i, row in enumerate(rows[:100]):
            try:
                name = row.find_element(By.XPATH, ".//h3").text
                
                link_el = row.find_element(By.XPATH, ".//a[contains(@href, '/channel/UC')]")
                href = link_el.get_attribute("href")
                channel_id = href.split("/channel/")[1] if "/channel/" in href else "ID 찾음 실패"
                
                youtube_link = f"https://youtube.com/channel/{channel_id}" if channel_id != "ID 찾음 실패" else ""

                tag_elements = row.find_elements(By.XPATH, ".//ul/li")
                tags = ", ".join([t.text for t in tag_elements if "#" in t.text])
                
                spans = row.find_elements(By.XPATH, ".//span[contains(@class, 'fluc-label')]")
                revenue = 0
                chat_count = 0
                
                if len(spans) >= 1:
                    clean_rev = re.sub(r'[^0-9]', '', spans[0].text)
                    revenue = int(clean_rev) if clean_rev else 0
                if len(spans) >= 2:
                    clean_cnt = re.sub(r'[^0-9]', '', spans[1].text)
                    chat_count = int(clean_cnt) if clean_cnt else 0

                # 컬럼
                result_list.append({
                    "수집월": date_str,
                    "순위": i + 1,
                    "채널명": name,
                    "태그": tags,              
                    "채널 ID": channel_id,
                    "유튜브 링크": youtube_link, 
                    "슈퍼챗 수입": revenue,
                    "슈퍼챗 개수": chat_count
                })
                
            except Exception as e:
                continue
        
        print(f"[{date_str}] 수집 완료")
        time.sleep(5) 

    # 4. 최종 데이터프레임 저장
    df = pd.DataFrame(result_list)
    filename = "v_tuber_superchat_202501_202603.csv"
    df.to_csv(filename, index=False, encoding="utf-8-sig")
    
    print("-" * 30)
    print(f"총 {len(df)}행 데이터 수집 및 저장 완료")
    print(f"파일명: {filename}")

except Exception as e:
    print(f"크롤링 중 치명적 에러 발생: {e}")
    # 에러나도 일단 저장하기
    if len(result_list) > 0:
        pd.DataFrame(result_list).to_csv("emergency_save.csv", index=False, encoding="utf-8-sig")
        print("중단점까지의 데이터를 emergency_save.csv로 저장했습니다.")

finally:
    driver.quit()
```

# 2025년, 2026년 스트리머 슈퍼챗 구하기

### 연간기준 2025 데이터 수집


```python
import pandas as pd
import re
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

# 1. 브라우저 설정
chrome_options = Options()
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

result_list = []

try:
    url = "https://playboard.co/youtube-ranking/most-superchatted-v-tuber-channels-in-south-korea-yearly?period=1735689600"
    
    print("2025년 연간 데이터를 수집합니다...")
    driver.get(url)
    time.sleep(40) 

    for _ in range(8):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)

    rows = driver.find_elements(By.XPATH, "//tr[descendant::h3]") or driver.find_elements(By.XPATH, "//ul/li[descendant::h3]")

    for i, row in enumerate(rows[:200]):
        try:
            name = row.find_element(By.XPATH, ".//h3").text
            channel_id = row.find_element(By.XPATH, ".//a[contains(@href, '/channel/UC')]").get_attribute("href").split("/channel/")[1]
            
            youtube_link = f"https://youtube.com/channel/{channel_id}"

            tag_elements = row.find_elements(By.XPATH, ".//ul/li")
            tags = ", ".join([t.text for t in tag_elements if "#" in t.text])
            
            spans = row.find_elements(By.XPATH, ".//span[contains(@class, 'fluc-label')]")
            revenue = int(re.sub(r'[^0-9]', '', spans[0].text)) if len(spans) >= 1 else 0
            chat_count = int(re.sub(r'[^0-9]', '', spans[1].text)) if len(spans) >= 2 else 0

            result_list.append({
                "연도": "2025",
                "순위": i + 1,
                "채널명": name,
                "태그": tags,
                "채널 ID": channel_id,
                "유튜브 링크": youtube_link,
                "슈퍼챗 수입": revenue,
                "슈퍼챗 개수": chat_count
            })
        except:
            continue

    # 3. 파일 저장
    df = pd.DataFrame(result_list)
    df.to_csv("annual_superchat_2025_official.csv", index=False, encoding="utf-8-sig")
    print("2025년 연간 데이터 수집 성공 (annual_superchat_2025_official.csv)")

finally:
    driver.quit()
```

### 2026 데이터는 월간으로 1월~3월 구해서 통합하기

```python
import pandas as pd

# 이전에 구했던 데이터에서 합치기
df = pd.read_csv("v_tuber_superchat_202501_202603.csv")

# 2026년만 빼오기
df_2026_monthly = df[df['수집월'].str.startswith('2026')]

# 채널 ID 기준으로 그룹화해서 수입과 개수를 모두 합산
annual_2026 = df_2026_monthly.groupby('채널 ID').agg({
    '채널명': 'first',       
    '태그': 'first',         
    '유튜브 링크': 'first',  
    '슈퍼챗 수입': 'sum',    # 1~3월 수입 다 더하기
    '슈퍼챗 개수': 'sum'     # 1~3월 개수 다 더하기
}).reset_index()

# 4. 내림차순 정렬
annual_2026 = annual_2026.sort_values('슈퍼챗 수입', ascending=False)

# 5. 순위 다시 매기기
annual_2026['순위'] = range(1, len(annual_2026) + 1)
annual_2026.insert(0, '연도', '2026') # 연도 컬럼 추가

# 6. 보기 좋게 컬럼 순서 맞추기
annual_2026 = annual_2026[['연도', '순위', '채널명', '태그', '채널 ID', '유튜브 링크', '슈퍼챗 수입', '슈퍼챗 개수']]

# 7. 최종 파일 저장
annual_2026.to_csv("annual_superchat_2026_official.csv", index=False, encoding="utf-8-sig")

print(f"2026년 (1~3월 합산) 연간 데이터 생성 완료! 총 {len(annual_2026)}명")
```

# 2025 + 2026 데이터 합치기

```python
import pandas as pd

try:
    df_2025 = pd.read_csv("annual_superchat_2025_official.csv")
    df_2026 = pd.read_csv("annual_superchat_2026_official.csv")
    
    print("파일을 성공적으로 불러왔습니다.")
    print(f"2025년 데이터: {len(df_2025)}행")
    print(f"2026년 데이터: {len(df_2026)}행")

    # 2. 위아래로 깔끔하게 합치기
    # ignore_index=True를 넣어야 0번부터 순서대로 인덱스가 예쁘게 다시 붙습니다.
    total_df = pd.concat([df_2025, df_2026], ignore_index=True)

    # 3. 최종 파일 저장하기
    final_filename = "v_tuber_annual_total_2025_2026.csv"
    total_df.to_csv(final_filename, index=False, encoding="utf-8-sig")

    print("-" * 30)
    print("병합 완료")
    print(f"총 합산 데이터 개수: {len(total_df)}행")
    print(f"저장된 파일명: {final_filename}")

except FileNotFoundError as e:
    print(f"에러 발생: 파일을 찾을 수 없습니다. 파일 이름을 다시 확인해 주세요!\n({e})")
```
