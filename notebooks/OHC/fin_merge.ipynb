```python
import pandas as pd
import re

# 선샌니의 3가지 규칙을 담은 마법의 전처리 함수!
def clean_streamer_name(name):
    # 만약 이름이 비어있다면(NaN) 그대로 패스!
    if pd.isna(name):
        return name
    
    name = str(name)
    
    # 규칙 1: 특수문자 무조건 제거 (한글, 영어, 숫자, 띄어쓰기만 남겨라!)
    name = re.sub(r'[^가-힣a-zA-Z0-9\s]', '', name)
    
    # 한글이 단 한 글자라도 포함되어 있는지 검사!
    if re.search(r'[가-힣]', name):
        # 규칙 2: 한글이 있으면 영문은 싹 지워버려라!
        name = re.sub(r'[a-zA-Z]', '', name)
    # 규칙 3: 한글이 없으면(영문만 있으면) 위 조건문을 안 타니까 영문 그대로 남음!
    
    # 마무리 단장: 너저분하게 남은 띄어쓰기들 깔끔하게 정리 (양끝 공백 싹둑!)
    name = re.sub(r'\s+', ' ', name).strip()
    
    return name

# 데이터 불러오기 (혹시 이미 df에 fancafe_final이 있다면 read_csv는 빼셔도 돼요!)
df_fancafe = pd.read_csv('fancafe_final.csv')

# 기존 'streamer_name' 열을 깨끗하게 씻겨서 다시 'streamer_name' 자리에 덮어쓰기! 드가자~
df_fancafe['streamer_name'] = df_fancafe['streamer_name'].apply(clean_streamer_name)

df_fancafe.to_csv('fancafe_final.csv', index=False, encoding='utf-8-sig')
```

```python
import pandas as pd
import numpy as np

# 1. 데이터 불러오기
df = pd.read_csv('final_poongtoday_2385.csv')

# 2. '월별 방송 시간' 역산하기 (이건 아까랑 똑같아요!)
df['월별_방송시간'] = np.where(df['hourly_star'] > 0, df['total_star'] / df['hourly_star'], 0)

# 3. 🌟 여기가 핵심! 이름(name)과 아이디(channel_id)를 둘 다 써서 묶어줍니다!
# 이렇게 하면 아이디가 사라지지 않고 옆에 딱 붙어있게 돼요.
df_grouped = df.groupby(['name', 'channel_id']).agg(
    총_별풍선=('total_star', 'sum'),
    총_방송시간=('월별_방송시간', 'sum')
).reset_index()

# 4. 합산 데이터로 '시간당 별풍선' 다시 계산하기
df_grouped['시간당_별풍선_평균'] = np.where(
    df_grouped['총_방송시간'] > 0, 
    df_grouped['총_별풍선'] / df_grouped['총_방송시간'], 
    0
)

# 5. 정수형(Int64)으로 깔끔하게 정리하기
df_grouped['총_별풍선'] = df_grouped['총_별풍선'].astype('Int64')
df_grouped['시간당_별풍선_평균'] = df_grouped['시간당_별풍선_평균'].round().astype('Int64')

# 6. 중간 계산용 '총_방송시간'은 이제 안녕~
df_final = df_grouped.drop(columns=['총_방송시간'])

# 7. 1줄로 압축된 데이터를 새로운 파일로 저장! (나중에 합칠 때 이 파일을 쓰면 돼요!)
df_final.to_csv('final_poongtoday_2385.csv', index=False, encoding='utf-8-sig')

# 잘 압축됐는지, ID는 잘 살아있는지 확인!
df_final.head()
```

```python
import pandas as pd
import numpy as np

# 1. 뼈대가 될 소프트콘 V2 데이터 불러오기
# (선샌니가 계속 덮어쓰기 하셨던 그 파일!)
df_softcone = pd.read_csv('final_softcone_streamers_V2.csv')

# 2. 1줄로 예쁘게 압축해둔 풍투데이 데이터 불러오기
# (아이디가 살아있는 최신 버전!)
df_poong = pd.read_csv('final_poongtoday_2385.csv')

# 3. 대공사 시작! channel_id를 기준으로 합치기 (LEFT JOIN)
# 소프트콘 목록은 그대로 유지하고, 풍투데이에서 [channel_id, 총_별풍선, 시간당_별풍선_평균]만 빌려옵니다.
df_merged = pd.merge(
    df_softcone, 
    df_poong[['channel_id', '총_별풍선', '시간당_별풍선_평균']], 
    on='channel_id', 
    how='left'
)

# 4. 빌려온 값들을 우리가 미리 만들어둔 '빈칸' 컬럼에 쏙쏙 이사시키기
df_merged['별풍선 도네이션'] = df_merged['총_별풍선']
df_merged['시간당 별풍선'] = df_merged['시간당_별풍선_평균']

# 5. 이사가 끝났으니 임시로 썼던 기둥('총_별풍선', '시간당_별풍선_평균')은 철거!
df_merged.drop(columns=['총_별풍선', '시간당_별풍선_평균'], inplace=True)

# 6. 마무리 청소: 소수점 떼고 깔끔하게 정수(Int64)로 변환
df_merged['별풍선 도네이션'] = df_merged['별풍선 도네이션'].astype('Int64')
df_merged['시간당 별풍선'] = df_merged['시간당 별풍선'].astype('Int64')

# 7. 완성된 데이터를 다시 V2에 덮어쓰기! 쾅쾅!
df_merged.to_csv('final_softcone_streamers_V2.csv', index=False, encoding='utf-8-sig')

# 결과가 얼마나 예쁜지 상위 5명만 스윽 볼까요?
df_merged[['streamer_name', 'channel_id', '별풍선 도네이션', '시간당 별풍선']].head()
```
